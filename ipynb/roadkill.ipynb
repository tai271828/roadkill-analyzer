```python
import pandas as pd
```

```python
csv_file = "https://www.freeway.gov.tw/UserFiles/File/Construction/opendata/roadkill.csv"
roadkill = pd.read_csv(csv_file,encoding = 'big5')
roadkill.head()
```

```python
filter = ["國道編號", "國道里程"]
roadkill_freeway_mile = roadkill[filter]
print(roadkill_freeway_mile)
```

```python
filter = (roadkill_freeway_mile["國道編號"] == 1.0)
roadkill_freewayone_mile = roadkill_freeway_mile[filter]
print(roadkill_freewayone_mile)
```

```python
filter1 = (roadkill_freeway_mile["國道編號"] == 1.0)
filter2 = (roadkill_freeway_mile["國道里程"] > -1.0)
filter = filter1 & filter2
roadkill_freeway_mile = roadkill[filter]
print(roadkill_freeway_mile)
```
